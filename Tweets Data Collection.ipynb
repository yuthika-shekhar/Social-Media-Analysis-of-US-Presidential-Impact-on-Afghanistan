### Fetching the data and dumping the results to JSON file

```python
def create_headers(bearer_token):
    headers = {
        "Authorization": "Bearer {}".format(bearer_token),
        "User-Agent": "v2SpacesSearchPython"
    }
    return headers


def connect_to_endpoint(url, headers, params):
    response = requests.request("GET", search_url, headers=headers, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()



headers = create_headers(bearer_token)
ctr = 0
for i in search_term:
    ctr = 0
    sleep(10)
    while(1):
        
        print(i)
        search_query = i + search_options   
        output_file_name = "Search_Term_"+i+"_JSON_data_file_"+str(ctr)+".json"
        output_file_path = project_dir+output_dir+output_file_name
        print(ctr)
        if ctr>0:
            query_params = {'query': search_query, 'start_time':'2020-11-02T00:00:00.0Z','end_time':'2020-12-02T00:00:00.0Z','tweet.fields':'author_id,created_at,geo,lang,public_metrics,source','expansions':'geo.place_id','place.fields':'country,full_name,place_type','user.fields':'created_at,description,name,location,public_metrics,username,url','max_results':500,'next_token':nt}
        else:
            query_params = {'query': search_query, 'start_time':'2020-11-02T00:00:00.0Z','end_time':'2020-12-02T00:00:00.0Z','tweet.fields':'author_id,created_at,geo,lang,public_metrics,source','expansions':'geo.place_id','place.fields':'country,full_name,place_type','user.fields':'created_at,description,name,location,public_metrics,username,url','max_results':500}
        json_response = connect_to_endpoint(search_url, headers, query_params)
        with open(output_file_path,'w',encoding='utf-8') as f:
            json.dump(json_response, f, ensure_ascii=False, indent=4)
        #print(x)
        ctr = ctr+1
        meta_data = json_response["meta"]
        if "next_token" in meta_data.keys():
            nt = json_response["meta"]["next_token"]
            sleep(10)
        else: 
            break
    
#if __name__ == "__main__":
#    main()
```

### Flattening the JSON file and save the results in CSV file

```python
filelist = listdir(output_dir)
ctr = 1
event_description = "U.S. Announces Troop Withdrawal"
for i in filelist:
    try:
        print(i)
        file_path = input_dir + i 
        term_start = i.split("Term_")[1]
        search_term = term_start.split("_")[0]
        with open(file_path,encoding = "utf-8") as jsonfile: 
            dict_train = json.load(jsonfile)

        if ctr==1:
            df = pd.json_normalize(dict_train["data"])
            nrows = df.shape[0]
            df["Search_Term"] = [search_term]*nrows
            df["Event_Title"] = [event_description]*nrows
            df["Filename"] = [i]*nrows
            ctr = ctr+1
        else: 
            inter_df = pd.json_normalize(dict_train["data"])
            nrows = inter_df.shape[0]
            inter_df["Search_Term"] = [search_term]*nrows
            inter_df["Event_Title"] = [event_description]*nrows
            inter_df["Filename"] = [i]*nrows
            df = df.append(inter_df, ignore_index=True)
            ctr = ctr+1
    except Exception as e:
        print("Error Occured")
        print(e)
try:
    #out_file_name = "Event_2_Tweets.csv" 
    #out_file_path = output_dir + out_file_name
    #df.to_csv(out_file_path)
    df
except Exception as e:
    print("Writing Error Occured")
    print(e)
    
```
